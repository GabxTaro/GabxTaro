#2. PEGAR LA COORDINACION AL DATAFRAME 234NA

```python
# Pegamos la columna Coordinacion, desde el dataframe MATERIAS
df234na_coord= pd.merge(df234na, materias[['CLAVE_UEA', 'COORDINACION']], left_on='UEA', right_on='CLAVE_UEA', how='left')

# Eliminamos la columna duplicada (CLAVE_UEA)
df234na_coord = df234na_coord.drop(['CLAVE_UEA','TRIM','NOMBRE','EDO','TRI1','TE1','TRI2','TE2','TRI3','TE3','TRI4','TE4','TRI5','TE5'], axis=1)

#Eliminamos la columna trimestre
#df234na_coord = df234na_coord.drop('TRIM', axis=1)

# Desplegamos una muestra del dataframe para verificar
df234na_coord.head()
```
